## We can remove NA with dropna(how) , here how specifies that only those rows would be removed who have any data except NA.

```python
# All NAN removed
nba.dropna()
```

```python
# Using how function , it will contain all the values where only one data is NAN.
nba.dropna(how="all")
```

```python
# WE can also drop Null values from Salary column
nba.dropna(subset = ["Salary"])
```

## Filling up NA 

```python
nba.fillna(0)
```

```python
# Filling 0 salary in place of NA and making it permanent
nba["Salary"].fillna(0,inplace=True)
```

```python
nba.head()
```

```python
nba["College"].fillna("None",inplace=True)
```

# Now changing the datatype of the dataframe , using astype()

```python
nba.dtypes
```

```python
nba.info()
```

```python
# astype has no inplace attribute
nba["Salary"] = nba["Salary"].astype("int")

# Its necessary to chaneg all NAN values 
nba["Number"].fillna(0,inplace=True)
nba["Age"].fillna(0,inplace=True)
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
```
