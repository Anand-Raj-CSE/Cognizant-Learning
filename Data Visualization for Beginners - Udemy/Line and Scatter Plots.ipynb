```python
import matplotlib.pyplot as plt
```

```python
with open("Goals.txt","r") as GoalData:
    homeTeamLine = GoalData.readline()
    homeTeamLine = homeTeamLine.strip(" \n")
    homeTeamLine = homeTeamLine.split(" ")
    homeTeamLine = [int(x) for x in homeTeamLine]
    
    awayTeamLine = GoalData.readline()
    awayTeamLine = awayTeamLine.strip(" \n")
    awayTeamLine = awayTeamLine.split(" ")
    awayTeamLine = [int(x) for x in awayTeamLine]
    

print(homeTeamLine)
print(awayTeamLine)
        
```

```python
plt.scatter(x=homeTeamLine,y=awayTeamLine)
plt.show()
```

```python
plt.scatter(x=homeTeamLine,y=awayTeamLine,s=70,c=[[1,0,0]]) # Adding RGB
plt.show()
```

```python
# Making every color value having its own color
plt.scatter(x=homeTeamLine,y=awayTeamLine,s=70,c=homeTeamLine)
plt.show()
```

```python
# Making every color value having its own color , having the distance from 0
import math
def distfromzero(htg,atg):
    return math.sqrt(htg**2+atg**2)

colors = []
for i in range(len(homeTeamLine)):
    colors.append(distfromzero(homeTeamLine[i],awayTeamLine[i]))

plt.scatter(x=homeTeamLine,y=awayTeamLine,s=70,c=colors)
plt.show()
```

```python
# We can use marker and use alpha for transparency
import math
def distfromzero(htg,atg):
    return math.sqrt(htg**2+atg**2)

colors = []
for i in range(len(homeTeamLine)):
    colors.append(distfromzero(homeTeamLine[i],awayTeamLine[i]))

plt.scatter(x=homeTeamLine,y=awayTeamLine,s=70,c=colors,marker = '*',alpha= 0.3)
plt.show()
```

```python
fig = plt.figure(figsize=(10,6)) # Width and height
plt.scatter(homeTeamLine,awayTeamLine,s=70,c=[[1,0.2,.3]],marker='*',alpha=0.3)
plt.show()
# This would create a figure of 10*6 , but lets see another example
```

```python
fig1 = plt.figure(figsize=(10,6)) # Width and height
fig2 = plt.figure(figsize=(6,10),facecolor="red")
plt.scatter(homeTeamLine,awayTeamLine,s=70,c=[[1,0.2,.3]],marker='*',alpha=0.3)
plt.show()
# This would create a figure of 6*10 as it is going to take the most recent one, we just put the plot over the figure
```
