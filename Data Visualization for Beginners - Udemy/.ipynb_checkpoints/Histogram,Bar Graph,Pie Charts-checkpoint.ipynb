```python
import matplotlib.pyplot as plt
```

```python
with open("Goals.txt","r") as GoalData:
    homeTeamLine = GoalData.readline()
    homeTeamLine = homeTeamLine.strip(" \n")
    homeTeamLine = homeTeamLine.split(" ")
    homeTeamGoals = [int(x) for x in homeTeamLine]
    
    awayTeamLine = GoalData.readline()
    awayTeamLine = awayTeamLine.strip(" \n")
    awayTeamLine = awayTeamLine.split(" ")
    awayTeamGoals = [int(x) for x in awayTeamLine]
    

print(homeTeamLine)
print(awayTeamLine)
```

```python
x=[]
for i in range(1,len(homeTeamLine)+1):
    x.append(i)
```

```python
fig = plt.figure("First",figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamLine)
```

### There is inconsistency in data as the bins are bigger and there are breaka too as the bins are by default set to 10.

```python
fig = plt.figure("First",figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamGoals,bins = 7 )#integers/sequence)
plt.show()
```

```python
fig = plt.figure("First",figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamGoals,bins=[0,1,2,3,4,5,6,7])
plt.show()
```
