```python
import matplotlib.pyplot as plt
```

```python
with open("Goals.txt","r") as GoalData:
    homeTeamLine = GoalData.readline()
    homeTeamLine = homeTeamLine.strip(" \n")
    homeTeamLine = homeTeamLine.split(" ")
    homeTeamGoals = [int(x) for x in homeTeamLine]
    
    awayTeamLine = GoalData.readline()
    awayTeamLine = awayTeamLine.strip(" \n")
    awayTeamLine = awayTeamLine.split(" ")
    awayTeamGoals = [int(x) for x in awayTeamLine]
    

print(homeTeamLine)
print(awayTeamLine)
```

```python
x=[]
for i in range(1,len(homeTeamLine)+1):
    x.append(i)
```

```python
fig = plt.figure("First",figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamLine)
```

### There is inconsistency in data as the bins are bigger and there are breaka too as the bins are by default set to 10.

```python
fig = plt.figure("First",figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamGoals,bins = 7 )#integers/sequence)
plt.show()
```

```python
fig = plt.figure("First",figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamGoals,bins=[0,1,2,3,4,5,6,7]) # This could also be done by range(8)
plt.show()
```

```python
# If we want ot draw histogram which has cummulative sums so we are going to just add a parameter cumulative
fig = plt.figure("First",figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.hist(x=homeTeamGoals,bins=[0,1,2,3,4,5,6,7],color = "Green",cumulative=True) 
plt.show()
```

```python
fig = plt.figure("First",figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.hist(x=(homeTeamGoals,awayTeamGoals),bins=[0,1,2,3,4,5,6,7],color = ("Blue","red"),cumulative=False) 
plt.show()
```

```python
# Each hist retuen 3 parameters
fig = plt.figure("First",figsize=(6,6))
ax = fig.add_subplot(1,1,1)
N, bins,patches = ax.hist(x=(homeTeamGoals,awayTeamGoals),bins=[0,1,2,3,4,5,6,7],color = ("Blue","red"),cumulative=False) 
plt.show()
```

```python
avgHomeTeam = sum(homeTeamGoals)/len(homeTeamGoals)
avgAwayTeam = sum(awayTeamGoals)/len(awayTeamGoals)
```

```python
ax.bar(left=[1,2], height=[avgHomeTeam,avgAwayTeam], width=0.35, align="center") # left is x cordinate, y cordinate is height
plt.show()
```
